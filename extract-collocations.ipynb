```python
def get_noun_or_verb(collocation):

    noun_lemma, verb_lemma = None, None
    for item in collocation:
        if item["upos"] == 'NOUN':
            noun_lemma = item['lemma']

        if item["upos"] == 'VERB':
            verb_lemma = item['lemma']
            
    if noun_lemma is not None:
        return noun_lemma
    
    if verb_lemma is not None:
        return verb_lemma
    
    return None
```

```python
words_dict = dict()
for item in result:
    if item[1] != 6:
        continue
    
    word = get_noun_or_verb(item[0])
    if word is None:
        continue
    if word in words_dict:
        words_dict[word] += 1
    else:
        words_dict[word] = 1    
```
