```python
import json
import os
import pandas as pd
from google.cloud import translate_v2 as translate
import time
```

```python
def load_json_file(file: str):
    with open(file, 'r') as f:
        d = json.load(f)
    return d    
```

```python
trip_advisor_file = '/Volumes/data/Data Science/Diploma/external_projects/data/Uzhhorod_Zakarpattia_Oblast.json'
```

```python
trip_advisior_list = load_json_file(trip_advisor_file)
```

```python
len(trip_advisior_list)
```

```python
trip_advisior_list[0]['rating']
```

```python
# Instantiates a client
translate_client = translate.Client()
target = 'uk'
source = 'ru'
```

```python
hotels, reviews_ru, reviews_uk, ratings = [], [], [], []
```

```python
length = 0
start = time.time()
for i in range(0, len(trip_advisior_list)):
    try:
        hotel = f"https://www.tripadvisor.ru{trip_advisior_list[i]['link']}"
        review_ru = trip_advisior_list[i]['text']
        length += len(review_ru)
        translated = translate_client.translate(review_ru, source_language=source, target_language=target)
        review_uk = translated['translatedText']
        rating = trip_advisior_list[i]['rating']
        
        hotels.append(hotel)
        reviews_ru.append(review_ru)
        reviews_uk.append(review_uk)
        ratings.append(rating)
        
        
    except Exception as e:
        print(e)
        delta = time.time() - start
        time.sleep(100 - delta)
        start = time.time()
        print(i)
        continue
# end = time.time()
# print(end - start)
```

```python
len(hotels)
```

```python
df = pd.DataFrame.from_dict({"hotel" : hotels, "review_ru" : reviews_ru, "review_uk": reviews_uk, "ratings" : ratings})
```
