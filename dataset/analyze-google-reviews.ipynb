```python
import json
```

```python
import langdetect
```

```python
def is_uk_language(text):
    try:
        if langdetect.detect(text) == 'uk':
            return True
    except:
        return False

    return False
```

```python
def load_json(file):
    with open(file, 'r') as f:
        data = json.load(f)
        return data
```

```python
def save_to_json(data, file):
    with open(file, 'w') as f:
        json.dump(data, f)
```

```python
path_dir = "google/kyiv"
```

```python
import os
files = []
for i in os.listdir(path_dir):
    if i.endswith('.json'):
        files.append(f"{path_dir}/{i}")
```

```python
path_dir2 = "google/lviv"
for i in os.listdir(path_dir2):
    if i.endswith('.json'):
        files.append(f"{path_dir2}/{i}")
```

```python
len(files)
```

```python
hotel_review_dict = dict()
for file in files:
    hotel_review_dict[file] = load_json(file)
```

```python
len(hotel_review_dict)
```

```python
%%time 
hotel_review_uk_text_dict = dict()
topic_text_dict = dict()
count_all_reviews = 0
count_reviews = 0
for hotel, review in hotel_review_dict.items():
#     print(review.keys())
    hotel_review_uk_text_dict[hotel] = []
    for item in review['reviews']:
        count_all_reviews += 1
        if item['should_be_more_details'] is False:
            if is_uk_language(item['text']):
                count_reviews += 1
                hotel_review_uk_text_dict[hotel].append(item['text'])
                for topic, text in item['topics'].items():
                    if text is not None:
                        if topic not in topic_text_dict:
                            topic_text_dict[topic] = [text]
                        else:
                            topic_text_dict[topic].append(text)
                    
```

```python
count_reviews
```

```python
sum_reviews = 0
for k,v in hotel_review_uk_text_dict.items():
    print(k)
    sum_reviews += len(v)
```
