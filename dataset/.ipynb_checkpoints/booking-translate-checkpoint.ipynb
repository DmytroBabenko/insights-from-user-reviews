```python
df = pd.read_csv('/Volumes/data/Data Science/Diploma/Detect-emotion-sentimental/dataset/ivano-frankivsk-reviews-ru.csv')
```

```python
df.head()
```

```python
df['title_uk'] = ''
df['pos_text_uk'] = ''
df['neg_text_uk'] = ''
```

```python
df.head()
```

```python
def is_correct_text(text: str):
    if type(text) is not str:
        return False
    
    if len(text) == 0:
        return False
    
    return True
```

```python
def translate_ru_uk(translate_client, text_ru: str):
    translated = translate_client.translate(text_ru, source_language='ru', target_language='uk')
    text_uk = translated['translatedText']
    return text_uk
```

```python
translate_client = translate.Client()
```

```python
translate_ru_uk(translate_client, 'Хоть и остановился в новом корпусе, но - матрац - старый с...')
```

```python
start = time.time()
for i in range(0, len(df)):
    try:
        title_ru = df.iloc[i]['title']
        if is_correct_text(title_ru):
            title_uk = translate_ru_uk(translate_client, title_ru)
            df['title_uk'][i] = title_uk

        pos_text_ru = df.iloc[i]['pos_text']
        if is_correct_text(pos_text_ru):
            pos_text_uk = translate_ru_uk(translate_client, pos_text_ru)
            df['pos_text_uk'][i] = pos_text_uk

        neg_text_ru = df.iloc[i]['neg_text']
        if is_correct_text(neg_text_ru):
            neg_text_uk = translate_ru_uk(translate_client, neg_text_ru)
            df['neg_text_uk'][i] = neg_text_uk
    
    except Exception as e:
        print(e)
        delta = time.time() - start
        if delta < 100:
            time.sleep(100 - delta)
        start = time.time()
        print(i)
        continue
```
