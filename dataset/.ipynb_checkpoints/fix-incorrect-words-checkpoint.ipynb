```python
import pandas as pd
from word_fixer import WordFixer
from gensim.models import KeyedVectors
from gensim.models.keyedvectors import Word2VecKeyedVectors
import tokenize_uk
from nltk.tokenize.treebank import TreebankWordDetokenizer
```

```python
embbedings_file = "../embeddings/ubercorpus.cased.tokenized.word2vec.300d"
```

```python
word2vec = KeyedVectors.load_word2vec_format(embbedings_file)
```

```python
word_fixer = WordFixer(word2vec)
```

```python
detokenizer = TreebankWordDetokenizer()
```

```python
df = pd.read_csv('booking/kyiv-reviews.csv')
```

```python
df = df.fillna("")
```

```python
df['neg_text'].values[9] = "В день зазду на вулиці трошки потеплішало і адміністрація встановила опалення на мінімум. Холодно для дорослих не було, але ми були з маленькою дитиною і на прийшлось купити тепловентілятор, що правда адміністрація не заперечувала, щоб ми його використовували."
```

```python
def fix_incorrect_words_in_sentence(sentence: str, incorrect_fixed_dict: dict):
    count = 0
    words = tokenize_uk.tokenize_words(sentence)
    for i in range(0, len(words)):
        if len(words[i]) <= 1:
            continue
        if word_fixer.is_word_correct(words[i]):
            continue
            
        try: 
            fixed_word = word_fixer.fix(words[i])
            incorrect_fixed_dict[words[i]] = fixed_word
            
            if words[i].lower() != fixed_word:
                count += 1

            words[i] = fixed_word
            
        except:
#             print(f"Exception was thrown for: {words[i]}")
            continue
                
    if count == 0:
        return 0, sentence
    
                
    fixed_sentences = detokenizer.detokenize(words)
    
    return count, fixed_sentences
```

```python
def fix_booking_df(df, incorrect_fixed_dict: dict):
    count = 0
    for i in range(0, len(df)):
        t, df['title'].values[i] = fix_incorrect_words_in_sentence(df['title'].values[i], incorrect_fixed_dict)
        p, df['pos_text'].values[i] = fix_incorrect_words_in_sentence(df['pos_text'].values[i], incorrect_fixed_dict)       
        n, df['neg_text'].values[i] = fix_incorrect_words_in_sentence(df['neg_text'].values[i], incorrect_fixed_dict)
        
        count += t + p + n
        
    return count
```

```python
def fix_booking_reviews(files):
    count = 0
    incorrect_fixed_dict = dict()
    for file in files:
        df = pd.read_csv(file)
        c = fix_booking_df(df, incorrect_fixed_dict)
        count += c
        
        df.to_csv(file, index=False)
        
    return count, incorrect_fixed_dict
```

### Fix words in original Ukrainian files

```python
original_uk_files = [
    'booking/kyiv-reviews.csv',
    'booking/lviv-reviews.csv',
    'booking/odesa-reviews.csv',
    'booking/kharkiv-reviews.csv',
    'booking/dnipro-reviews.csv',
    'booking/ivano-frankivsk-reviews.csv',
    'booking/uzhgorod-reviews.csv'
]
```

```python
count, incorrect_fixed_dict = fix_booking_reviews(original_uk_files)
```

```python
count
```

```python
incorrect_fixed_dict
```

```python
def count_all_booking_words(files):
    count = 0
    for file in files:
        df = pd.read_csv(file)
        
        title_words = sum(df['title'].apply(lambda x: len(str(x).split())).values)
        pos_words = sum(df['pos_text'].apply(lambda x: len(str(x).split())).values)
        neg_words = sum(df['neg_text'].apply(lambda x: len(str(x).split())).values)
        
        
        count += title_words + pos_words + neg_words  
    return count
```

```python
num_words = count_all_booking_words(original_uk_files)
```

```python
num_words
```

```python
print(f"There were fixed {count} of all {num_words} words, {100* count/num_words}%")
```

### Fix words in Tripadvisor 

```python
def fix_tripadvisor_df(df, incorrect_fixed_dict: dict):
    count = 0
    for i in range(0, len(df)):
        c, df['review_uk'].values[i] = fix_incorrect_words_in_sentence(df['review_uk'].values[i], incorrect_fixed_dict)
        count += c
        
    return count
```

```python
def fix_tripadvisor_reviews(files):
    count = 0
    incorrect_fixed_dict = dict()
    for file in files:
        df = pd.read_csv(file)
        c = fix_tripadvisor_df(df, incorrect_fixed_dict)
        count += c
        
        df.to_csv(file, index=False)
        
    return count, incorrect_fixed_dict
```

```python
tripadvisor_files = [
    'trip-advisor/tripadvisor-kyiv.csv',
    'trip-advisor/tripadvisor-lviv.csv',
    'trip-advisor/tripadvisor-odesa.csv',
    'trip-advisor/tripadvisor-kharkiv.csv',
    'trip-advisor/tripadvisor-dnipro.csv',
    'trip-advisor/tripadvisor-ivano_frankivsk.csv',
    'trip-advisor/tripadvisor-uzhhorod.csv'
]
```

```python
trip_advisor_count, tripadvisor_fixed_dict = fix_tripadvisor_reviews(tripadvisor_files)
```

```python
def count_all_tripadvisor_words(files):
    count = 0
    for file in files:
        df = pd.read_csv(file)
        
        words = sum(df['review_uk'].apply(lambda x: len(str(x).split())).values)
        
        count += words  
    return count
```

```python
num_trip_advisor_words = count_all_tripadvisor_words(tripadvisor_files)
```

```python
print(f"There were fixed {trip_advisor_count} of all {num_trip_advisor_words} words, {100* trip_advisor_count/num_trip_advisor_words}%")
```
