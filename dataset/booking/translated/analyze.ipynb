```python
import pandas as pd
```

```python
def read_data_frame_from_files(files):
    frames = []
    for file in files:
        df = pd.read_csv(file)
        frames.append(df)
        
    return pd.concat(frames)
```

```python
def remove_unnamed_column(files):
    for file in files:
        df = pd.read_csv(file)
        df.iloc[:,1:].to_csv(file, index=False)
```
