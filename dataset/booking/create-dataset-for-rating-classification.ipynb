```python
def convert10to5star(rate):
    stars = round(rate / 2)
    if stars < 1:
        stars = 1
        
    return stars
```

```python
def add_5_star_rating(file):
    df = pd.read_csv(file)
    df['rating'] = df['ratingValue'].apply(convert10to5star)
    df.to_csv(file, index=False)
```

```python
files = [
    'booking-train.csv',
    'booking-test.csv',
    'booking-val.csv'
]
```

```python
for file in files:
    add_5_star_rating(file)
```

```python
def create_rating_classification_dataset(file):
    df = pd.read_csv(file)
    texts = []
    for i in range(0, len(df)):
        title = df['title'].values[i]
        if type(title) is not str:
            title = ""
            
        pos_text = df['pos_text'].values[i]
        if type(pos_text) is not str:
            pos_text = ""

        neg_text = df['neg_text'].values[i]
        if type(neg_text) is not str:
            neg_text = ""
            
        text = f"{title} {pos_text} {neg_text}"
        
        texts.append(text)
        
    ratings = df['rating'].values
    assert len(texts) == len(ratings)
    
    return pd.DataFrame.from_dict({'text' : texts, 'rating' : ratings})
```

```python
booking_rating_train_df = create_rating_classification_dataset('booking-train.csv')
```

```python
booking_rating_train_df.to_csv('booking-rating-train.csv')
```

```python
booking_rating_test_df = create_rating_classification_dataset('booking-test.csv')
```

```python
booking_rating_test_df.to_csv('booking-rating-test.csv')
```

```python
booking_rating_val_df = create_rating_classification_dataset('booking-val.csv')
```

```python
booking_rating_val_df.to_csv('booking-rating-val.csv')
```

### dataset for one hot hack

```python
hack_word = '123456789'
```

```python
def create_rating_classification_dataset_for_one_hot(file):
    df = pd.read_csv(file)
    texts = []
    for i in range(0, len(df)):
        title = df['title'].values[i]
        if type(title) is not str:
            title = ""
            
        pos_text = df['pos_text'].values[i]
        if type(pos_text) is not str:
            pos_text = ""

        neg_text = df['neg_text'].values[i]
        if type(neg_text) is not str:
            neg_text = ""
            
        text = f"{title} {hack_word} {pos_text} {hack_word} {neg_text}"
        
        texts.append(text)
        
    ratings = df['rating'].values
    assert len(texts) == len(ratings)
    
    return pd.DataFrame.from_dict({'text' : texts, 'rating' : ratings})
```

```python
booking_rating_for_one_hot_train_df = create_rating_classification_dataset_for_one_hot('booking-train.csv')
```

```python
booking_rating_for_one_hot_train_df.to_csv('booking-rating-for-one-hot-train.csv')
```

```python
booking_rating_for_one_hot_test_df = create_rating_classification_dataset_for_one_hot('booking-test.csv')
```

```python
booking_rating_for_one_hot_test_df.to_csv('booking-rating-for-one-hot-test.csv')
```

```python
booking_rating_for_one_hot_val_df = create_rating_classification_dataset_for_one_hot('booking-val.csv')
```

```python
booking_rating_for_one_hot_val_df.to_csv('booking-rating-for-one-hot-val.csv')
```

```python
def rename_rating_to_label(file):
    df = pd.read_csv(file)
    renamed_df = df.rename(columns={"rating": "label"})
    renamed_df.to_csv(file, index=False)
```

```python
files = [
    'booking-rating-train.csv',
    'booking-rating-test.csv',
    'booking-rating-val.csv',
    'booking-rating-for-one-hot-train.csv',
    'booking-rating-for-one-hot-test.csv',
    'booking-rating-for-one-hot-val.csv'
]
```

```python
for file in files:
    rename_rating_to_label(file)
```
