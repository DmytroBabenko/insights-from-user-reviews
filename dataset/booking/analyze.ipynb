```python
import pandas as pd
```

```python
def read_data_frame_from_files(files):
    frames = []
    for file in files:
        df = pd.read_csv(file)
        frames.append(df)
        
    return pd.concat(frames)
```

#### Original data

```python
original_uk_fils = [
    'kyiv-reviews.csv',
    'lviv-reviews.csv',
    'odesa-reviews.csv',
    'kharkiv-reviews.csv',
    'dnipro-reviews.csv',
    'ivano-frankivsk-reviews.csv',
    'uzhgorod-reviews.csv'
]
```

```python
original_uk_df = read_data_frame_from_files(original_uk_fils)
```

```python
translted_to_uk_files = [
    'kyiv-reviews-translated.csv',
    'lviv-reviews-translated.csv',
    'odesa-reviews-translated.csv',
    'kharkiv-reviews-translated.csv',
    'dnipro-reviews-translated.csv',
    'ivano-frankivsk-reviews-translated.csv',
    'uzhgorod-reviews-translated.csv'
]
```
