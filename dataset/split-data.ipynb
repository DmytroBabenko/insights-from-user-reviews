### Booking.com

```python
def read_all_df(files):
    frames = []
    for file in files:
        df = pd.read_csv(file)
        print(file)
        df = df[['title', 'pos_text', 'neg_text', 'ratingValue', 'bestRating', 'hotel']]
        frames.append(df)
    
    return pd.concat(frames)
```

```python
def read_trip_advisor_df(files):
    frames = []
    for file in files:
        df = pd.read_csv(file)
        frames.append(df)
    
    return pd.concat(frames)
```

```python
trip_advisor_files = [
    'trip-advisor/tripadvisor-kyiv.csv',
    'trip-advisor/tripadvisor-lviv.csv',
    'trip-advisor/tripadvisor-odesa.csv',
    'trip-advisor/tripadvisor-kharkiv.csv',
    'trip-advisor/tripadvisor-dnipro.csv',
    'trip-advisor/tripadvisor-ivano_frankivsk.csv',
    'trip-advisor/tripadvisor-uzhhorod.csv',
]
```

```python
trip_advisor_df = read_trip_advisor_df(trip_advisor_files)
```

```python
trip_advisor_df
```

```python
np.unique(trip_advisor_df['hotel'].values).shape
```

```python
def split_data(df):
    unique_hotels = np.unique(df['hotel'].values)
    np.random.shuffle(unique_hotels)
    
    train_last_idx = int(0.6 * len(unique_hotels))
    test_last_idx = int(0.8 * len(unique_hotels))
    
    train_hotels = unique_hotels[:train_last_idx]
    test_hotels = unique_hotels[train_last_idx:test_last_idx]
    val_hotels = unique_hotels[test_last_idx:]
    
    train_df = df.loc[df['hotel'].isin(train_hotels)]
    test_df = df.loc[df['hotel'].isin(test_hotels)]
    val_df = df.loc[df['hotel'].isin(val_hotels)]
    
    return train_df, test_df, val_df
    
```

```python
trip_advisor_train_df, trip_advisor_test_df, trip_advisor_val_df = split_data(trip_advisor_df)
```

```python
len(trip_advisor_train_df) / len(trip_advisor_df)
```

```python
len(trip_advisor_test_df) / len(trip_advisor_df)
```

```python
len(trip_advisor_val_df) / len(trip_advisor_df)
```

```python
trip_advisor_train_df.to_csv("trip-advisor/trip-advisor-train.csv", index=False)
```

```python
trip_advisor_test_df.to_csv("trip-advisor/trip-advisor-test.csv", index=False)
```

```python
trip_advisor_val_df.to_csv("trip-advisor/trip-advisor-val.csv", index=False)
```
